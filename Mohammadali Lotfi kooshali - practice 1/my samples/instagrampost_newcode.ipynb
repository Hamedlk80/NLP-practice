# install , download and import libraries

```python
!pip install spacy nltk
!python -m spacy download en_core_web_sm

import re
import spacy
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer, SnowballStemmer
from nltk import pos_tag
from nltk.corpus import wordnet as wn
from string import punctuation
from nltk.chunk import RegexpParser
from nltk import Tree
import time

# Download required NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('omw-1.4')

# Load spaCy model
nlp = spacy.load("en_core_web_sm")
```

# reading text file

```python
with open('instagram_post.txt', 'r', encoding='utf-8') as file:
    raw_text = file.read()

print("original text:")
print(raw_text)
print(f"Total characters: {len(raw_text)}")
```

# 1. Cleaning Text with re.sub


```python
# Remove numbers
text_no_numbers = re.sub(r'\d+', '', raw_text)
print("After removing numbers:")
print(text_no_numbers)
print("-" * 80)

# Remove extra whitespaces
text_cleaned = re.sub(r'\s+', ' ', text_no_numbers)
print("After removing extra whitespaces:")
print(text_cleaned)
print("-" * 80)

# Remove special characters (keep letters, spaces, dots, commas)
text_cleaned = re.sub(r'[^\w\s\.\,]', '', text_cleaned)
print("After removing special characters:")
print(text_cleaned)
print(f"Cleaned text length: {len(text_cleaned)} characters")
```

# 2. Lowercasing

```python
text_lower = text_cleaned.lower()
print("Text after lowercasing:")
print(text_lower)
```

# 3. Punctuation Removal

```python
print("Punctuation removed:", punctuation)
text_no_punct = ''.join([char for char in text_lower if char not in punctuation])
print("Text after punctuation removal:")
print(text_no_punct)
```

# 4. Stopword Removal

```python
# Get English stopwords
stop_words = set(stopwords.words('english'))
print("English stopwords :")
print(list(stop_words))

# Tokenize first
tokens = word_tokenize(text_no_punct)
print(f"Total tokens before stopword removal: {len(tokens)}")
print("tokens before removal:")
print(tokens)

# Remove stopwords
filtered_tokens = [word for word in tokens if word not in stop_words]
print(f"Total tokens after stopword removal: {len(filtered_tokens)}")
print("tokens after removal:")
print(filtered_tokens)

# Convert back to text
text_no_stopwords = ' '.join(filtered_tokens)
print("Complete text after stopword removal:")
print(text_no_stopwords)
```

# 5. Tokenization (Sentence, Word, Character)

```python
# Sentence Tokenization
sentences = sent_tokenize(text_lower)
print("Sentence Tokenization:")
print(f"Total number of sentences: {len(sentences)}")
print("All sentences:")
for i, sent in enumerate(sentences, 1):
    print(f"{i}. {sent}")

# Word Tokenization
words = word_tokenize(text_lower)
print("Word Tokenization:")
print(f"Total number of words: {len(words)}")
print("All words:")
print(words)

# Character Tokenization
chars = [char for char in text_lower if char != ' ']
print("Character Tokenization:")
print(f"Total number of characters (without spaces): {len(chars)}")
print("All characters:")
print(chars)
```

# 6. Stemming (PorterStemmer and SnowballStemmer)

```python
# Initialize both stemmers
porter_stemmer = PorterStemmer()
snowball_stemmer = SnowballStemmer("english")

# Apply both stemmers
porter_stemmed_tokens = [porter_stemmer.stem(word) for word in filtered_tokens]
snowball_stemmed_tokens = [snowball_stemmer.stem(word) for word in filtered_tokens]

print("COMPLETE COMPARISON (Original vs Porter vs Snowball):")
print(f"{'#':<5} | {'Original Word':<25} | {'Porter Stemmer':<20} | {'Snowball Stemmer':<20} | {'Difference'}")

difference_count = 0
for i, (original, porter, snowball) in enumerate(zip(filtered_tokens, porter_stemmed_tokens, snowball_stemmed_tokens), 1):
    difference = "DIFFERENT" if porter != snowball else "SAME"
    if porter != snowball:
        difference_count += 1
    print(f"{i:<5} | {original:<25} | {porter:<20} | {snowball:<20} | {difference}")

print(f"\nTotal words with different stemming results: {difference_count} out of {len(filtered_tokens)}")
print(f"Percentage of differences: {(difference_count/len(filtered_tokens))*100:.2f}%")

# Show detailed comparison for words that differ
print("\nDETAILED COMPARISON FOR WORDS WITH DIFFERENT STEMMING RESULTS:")

different_words = []
for original, porter, snowball in zip(filtered_tokens, porter_stemmed_tokens, snowball_stemmed_tokens):
    if porter != snowball:
        different_words.append((original, porter, snowball))

if different_words:
    print(f"{'#':<5} | {'Original Word':<25} | {'Porter Result':<20} | {'Snowball Result':<20} | {'Analysis'}")
    for i, (original, porter, snowball) in enumerate(different_words, 1):
        analysis = ""
        if len(porter) < len(snowball):
            analysis = "Porter shorter"
        elif len(snowball) < len(porter):
            analysis = "Snowball shorter"
        elif porter.endswith('ing') and not snowball.endswith('ing'):
            analysis = "Porter keeps -ing"
        elif snowball.endswith('ing') and not porter.endswith('ing'):
            analysis = "Snowball keeps -ing"
        else:
            analysis = "Different algorithm"
        print(f"{i:<5} | {original:<25} | {porter:<20} | {snowball:<20} | {analysis}")
else:
    print("No differences found between Porter and Snowball stemmers for this text.")

# Show complete stemmed texts
print("\nCOMPLETE PORTER STEMMED TEXT:")
porter_stemmed_text = ' '.join(porter_stemmed_tokens)
print(porter_stemmed_text)

print("\nCOMPLETE SNOWBALL STEMMED TEXT:")
snowball_stemmed_text = ' '.join(snowball_stemmed_tokens)
print(snowball_stemmed_text)

# Performance comparison
print("\nPERFORMANCE COMPARISON:")

# Test Porter speed
start_time = time.time()
for word in filtered_tokens * 10:
    porter_stemmer.stem(word)
porter_time = time.time() - start_time

# Test Snowball speed
start_time = time.time()
for word in filtered_tokens * 10:
    snowball_stemmer.stem(word)
snowball_time = time.time() - start_time

print(f"PorterStemmer processing time (10 repetitions): {porter_time:.6f} seconds")
print(f"SnowballStemmer processing time (10 repetitions): {snowball_time:.6f} seconds")
```

# 7. Lemmatization (WordNetLemmatizer)

```python
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("COMPLETE COMPARISON (Original vs Lemmatized):")
print(f"{'#':<5} | {'Original Word':<25} | {'Lemmatized':<25}")
for i, (original, lemmatized) in enumerate(zip(filtered_tokens, lemmatized_tokens), 1):
    print(f"{i:<5} | {original:<25} | {lemmatized:<25}")

lemmatized_text = ' '.join(lemmatized_tokens)
print("\nCOMPLETE LEMMATIZED TEXT:")
print(lemmatized_text)

print("\nSTEMMING vs LEMMATIZATION COMPARISON:")
print(f"{'#':<5} | {'Original Word':<25} | {'Porter Stemmer':<20} | {'Snowball Stemmer':<20} | {'Lemmatized':<25}")
for i, (original, porter, snowball, lemmatized) in enumerate(zip(filtered_tokens, porter_stemmed_tokens, snowball_stemmed_tokens, lemmatized_tokens), 1):
    print(f"{i:<5} | {original:<25} | {porter:<20} | {snowball:<20} | {lemmatized:<25}")

# Performance comparison for Lemmatizer
print("\nPERFORMANCE COMPARISON (Lemmatizer vs Stemmers):")

# Test Lemmatizer speed
start_time = time.time()
for word in filtered_tokens * 10:
    lemmatizer.lemmatize(word)
lemmatizer_time = time.time() - start_time

print(f"WordNetLemmatizer processing time (10 repetitions): {lemmatizer_time:.6f} seconds")
print(f"PorterStemmer processing time (10 repetitions): {porter_time:.6f} seconds")
print(f"SnowballStemmer processing time (10 repetitions): {snowball_time:.6f} seconds")
```

# 8. Part-of-Speech (POS) Tagging

```python
# Cell 10: 8. POS Tagging (Part-of-Speech Tagging)
print("8. PART-OF-SPEECH (POS) TAGGING")

# Perform POS tagging on all filtered tokens
pos_tagged = pos_tag(filtered_tokens)

print("COMPLETE POS TAGGING RESULTS:")
print(f"{'#':<4} | {'Word':<20} | {'POS Tag':<10} | {'Description'}")

pos_descriptions = {
    'NN': 'Noun, singular', 'NNS': 'Noun, plural',
    'NNP': 'Proper noun, singular', 'NNPS': 'Proper noun, plural',
    'VB': 'Verb, base form', 'VBD': 'Verb, past tense',
    'VBG': 'Verb, gerund/present participle', 'VBN': 'Verb, past participle',
    'VBP': 'Verb, non-3rd person singular present', 'VBZ': 'Verb, 3rd person singular present',
    'JJ': 'Adjective', 'JJR': 'Adjective, comparative', 'JJS': 'Adjective, superlative',
    'RB': 'Adverb', 'RBR': 'Adverb, comparative', 'RBS': 'Adverb, superlative',
    'IN': 'Preposition/subordinating conjunction', 'DT': 'Determiner',
    'CC': 'Coordinating conjunction', 'PRP': 'Personal pronoun', 'PRP$': 'Possessive pronoun',
    'MD': 'Modal auxiliary', 'TO': 'to', 'UH': 'Interjection',
    'CD': 'Cardinal number', 'EX': 'Existential there', 'FW': 'Foreign word',
    'LS': 'List item marker', 'POS': 'Possessive ending', 'RP': 'Particle',
    'SYM': 'Symbol', 'WDT': 'Wh-determiner', 'WP': 'Wh-pronoun',
    'WP$': 'Possessive wh-pronoun', 'WRB': 'Wh-adverb'
}

for i, (word, tag) in enumerate(pos_tagged, 1):
    desc = pos_descriptions.get(tag, 'Other')
    print(f"{i:<4} | {word:<20} | {tag:<10} | {desc}")
```

# 9. Chunking

```python
# Define chunk grammar
chunk_grammar = r"""
    NP: {<DT>?<JJ>*<NN.*>+}
    NP: {<NNP>+}
    NP: {<NNP><POS>?<NNP>*}
    VP: {<VB.*><NP|PP|CLAUSE>+$}
    PP: {<IN><NP>}
"""
chunk_parser = RegexpParser(chunk_grammar)

# Process ALL sentences for chunking
all_sentences = sent_tokenize(text_lower)

for sent_idx, sentence in enumerate(all_sentences, 1):
    print(f"\nSentence {sent_idx}: {sentence}")

    # Tokenize and tag
    tokens = word_tokenize(sentence)
    pos_tags = pos_tag(tokens)

    # Parse chunks
    chunk_tree = chunk_parser.parse(pos_tags)

    print("Chunk Tree Structure:")
    print(chunk_tree)

    print("Chunks extracted:")
    for subtree in chunk_tree.subtrees():
        if subtree.label() in ['NP', 'VP', 'PP']:
            chunk_words = ' '.join(word for word, tag in subtree.leaves())
            print(f"  {subtree.label()}: {chunk_words}")
```

# 10. Parsing ( Dependency Parsing and Tree Structure Parsing )

```python
# Process entire text with spaCy
doc = nlp(text_lower)

print("DEPENDENCY RELATIONS FOR ENTIRE TEXT:")

for token in doc:
    print(f"Word: '{token.text}'")
    print(f"  Head word: '{token.head.text}'")
    print(f"  Dependency relation: {token.dep_}")
    print(f"  Explanation: {spacy.explain(token.dep_)}")
    print(f"  Children: {[child.text for child in token.children]}")
```

```python
all_sentences = sent_tokenize(text_lower)

for sent_idx, sentence in enumerate(all_sentences, 1):
    print(f"\nSentence {sent_idx}: {sentence}")

    # Tokenize and tag
    tokens = word_tokenize(sentence)
    pos_tags = pos_tag(tokens)

    # Create a simple binary tree structure for visualization
    def create_simple_tree(tokens, pos_tags):
        """Create a simple constituency-like tree structure"""
        if len(tokens) == 1:
            return Tree(pos_tags[0][1], [tokens[0]])
        else:
            mid = len(tokens) // 2
            left_tree = create_simple_tree(tokens[:mid], pos_tags[:mid])
            right_tree = create_simple_tree(tokens[mid:], pos_tags[mid:])
            return Tree('S', [left_tree, right_tree])

    try:
        constituency_tree = create_simple_tree(tokens, pos_tags)
        print("Constituency Tree:")
        constituency_tree.pretty_print()

        print("Tree in bracketed notation:")
        print(constituency_tree)
    except Exception as e:
        print(f"Error creating tree: {e}")

print("\nALTERNATIVE TREE VISUALIZATION (ALL SENTENCES - Detailed NP chunks):")

# More detailed tree structure for ALL sentences
for sent_idx, sentence in enumerate(all_sentences, 1):
    print(f"\nSentence {sent_idx}: {sentence}")

    tokens = word_tokenize(sentence)
    pos_tags = pos_tag(tokens)

    # Manual tree building
    tree = Tree('S', [])
    current_np = []

    for word, tag in pos_tags:
        if tag.startswith('NN') or tag.startswith('JJ') or tag == 'DT':
            current_np.append((word, tag))
        else:
            if current_np:
                np_tree = Tree('NP', [f"{w}/{t}" for w, t in current_np])
                tree.append(np_tree)
                current_np = []
            tree.append(f"{word}/{tag}")

    if current_np:
        np_tree = Tree('NP', [f"{w}/{t}" for w, t in current_np])
        tree.append(np_tree)

    print("Detailed Constituency Tree (NP chunks):")
    tree.pretty_print()
    print("\nTree as string:")
    print(tree)
```

# 11. Hypernyms and Hyponyms (WordNet)

```python
# Extract ALL unique words from filtered tokens
unique_words = list(set(filtered_tokens))

for word in unique_words:
    print(f"\nWORD: '{word}'")

    # Get synsets for the word
    synsets = wn.synsets(word)

    if not synsets:
        print(f"  No synsets found for '{word}'")
        continue

    # Take first synset
    synset = synsets[0]
    print(f"  Definition: {synset.definition()}")
    print(f"  Example: {synset.examples() if synset.examples() else 'No example'}")

    # Hypernyms (more general concepts)
    hypernyms = synset.hypernyms()
    if hypernyms:
        print(f"\n  HYPERNYMS (more general): {len(hypernyms)} found")
        for hyp in hypernyms:
            print(f"    → {hyp.name()}")
            print(f"      Definition: {hyp.definition()}")
            # Get path hypernyms (higher level)
            hyper_paths = hyp.hypernyms()
            if hyper_paths:
                for path_hyp in hyper_paths:
                    print(f"        ↑ {path_hyp.name()}: {path_hyp.definition()}")
    else:
        print(f"\n  HYPERNYMS: None found")

    # Hyponyms (more specific concepts)
    hyponyms = synset.hyponyms()
    if hyponyms:
        print(f"\n  HYPONYMS (more specific): {len(hyponyms)} found")
        for hyp in hyponyms:
            print(f"    → {hyp.name()}")
            print(f"      Definition: {hyp.definition()}")
    else:
        print(f"\n  HYPONYMS: None found")

    # Full hierarchy path (no depth limit)
    print(f"\n  HIERARCHY PATH (up to root):")
    def show_path(s, level=0):
        print(f"    {'  ' * level}└─ {s.name()}")
        for hyp in s.hypernyms():
            show_path(hyp, level + 1)

    show_path(synset)

# Detailed example for ALL specific words from the text
print("\nDETAILED EXAMPLE FOR KEY WORDS FROM TEXT")

key_words = ['apple', 'computer', 'company', 'inventor', 'founder', 'ceo', 'pixar', 'next']

for word in key_words:
    print(f"\nDEEP ANALYSIS FOR: '{word}'")

    synsets = wn.synsets(word)
    if synsets:
        primary_synset = synsets[0]

        print(f"  Definition: {primary_synset.definition()}")
        print(f"  Lemmas: {primary_synset.lemma_names()}")

        # Get all hypernym paths (no limit)
        print("\n  All Hypernym Paths:")
        hypernym_paths = primary_synset.hypernym_paths()
        for i, path in enumerate(hypernym_paths, 1):
            print(f"    Path {i}: {' -> '.join([s.name().split('.')[0] for s in path])}")

        # Get all hyponyms (no limit on display)
        all_hyponyms = primary_synset.hyponyms()
        print(f"\n  Total hyponyms found: {len(all_hyponyms)}")
        if all_hyponyms:
            print("  All hyponyms:")
            for hyp in all_hyponyms:
                print(f"    - {hyp.name()}: {hyp.definition()}")
```
